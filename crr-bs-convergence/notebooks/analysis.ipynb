## 1. Paper parameters (replication)

```python
params = OptionParams(
    S0=27.0, K=25.0, r=0.015, sigma=0.20, T=0.410959,
    option_type=OptionType.PUT,
    exercise_style=ExerciseStyle.EUROPEAN,
    param=Parametrisation.CRR,
    discounting=DiscountingMode.CONTINUOUS,
)

print(f'Black-Scholes put price : {black_scholes(params):.6f}')
print(f'CRR N=5 put price       : {crr_price(params, 5):.6f}')
print(f'CRR N=100 put price     : {crr_price(params, 100):.6f}')
```

```python
# Convergence table
data = convergence_series(params, n_steps=[1,2,3,5,10,20,50,100,200])
df = pd.DataFrame(data).set_index('N')
df.columns = ['CRR price', 'JR price', 'BS price', 'Error CRR', 'Error JR']
df.style.format('{:.6f}').background_gradient(subset=['Error CRR','Error JR'], cmap='Reds_r')
```

## 2. Convergence plot

```python
fig = convergence_plot(params)
fig.show()
```

## 3. Even/Odd oscillation

```python
parity_oscillation_plot(params).show()
```

## 4. Sensitivity analysis

```python
sensitivity_plot(params, vary='sigma').show()
sensitivity_plot(params, vary='S0').show()
```

## 5. Binomial stock tree (N=5)

```python
stock_tree_plot(params, N=5).show()
```
