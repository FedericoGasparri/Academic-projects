## Abstract

This paper analyses the Cox-Ross-Rubinstein (CRR) binomial option pricing model and its convergence to the Black-Scholes (BS) continuous-time limit. We introduce the model's mathematical structure, discuss both the CRR and Jarrow-Rudd (JR) parametrisations, and derive European and American option prices via risk-neutral valuation. The central result — established through the weak convergence of a normalised random walk to Brownian motion under the risk-neutral measure — is that CRR prices converge to BS prices as the number of time steps $n \to \infty$. The convergence is illustrated numerically and visually using an interactive implementation in Python.

---
## 1 · Introduction

The Cox-Ross-Rubinstein (CRR) model, introduced by Cox, Ross and Rubinstein (1979), is a discrete-time framework for option pricing that extends the principles of the Black-Scholes-Merton model. By representing the price of an underlying asset as a recombining binomial tree, the model offers an intuitive and computationally tractable approach to derivative valuation.

At each time step the asset price can move either up by a factor $u$ or down by a factor $d$. The option value is then determined by **backward induction**: starting from the known payoffs at maturity, the model discounts expected values to the present using a risk-neutral probability measure that precludes arbitrage.

The CRR model is particularly valued for its flexibility. Unlike closed-form continuous-time models, it naturally accommodates **American options** (where early exercise is optimal at certain nodes), dividend-paying stocks, and a variety of exotic payoffs. Its primary theoretical appeal, however, lies in its role as a discretisation of the Black-Scholes model: as the time step $\Delta t \to 0$, the binomial tree converges — in distribution — to the geometric Brownian motion that underpins Black-Scholes.

---
## 2 · The CRR Binomial Model

### 2.1 · Core Assumptions

The CRR model rests on the following assumptions:

- The asset price evolves in **discrete time steps** $t \in \{0, 1, \ldots, n\}$.
- At each node, the price moves to exactly **two states**: up by factor $u > 1$ or down by factor $d < 1$.
- The **risk-free rate** $r$ is constant over the life of the option.
- There are **no transaction costs, taxes, or dividends** (in the basic formulation).
- The market is **arbitrage-free and complete**, which requires $d < 1 + r < u$.

The completeness condition guarantees the existence of a **unique** risk-neutral measure $\mathbb{Q}$, so there is a unique fair price for every European contingent claim.

### 2.2 · Stock Price Tree

Starting from an initial price $S_0 > 0$, the stock price at time step $t$ after $j$ up-moves is:

$$S_t^\omega = S_0 \cdot u^{N_t(\omega)} \cdot d^{t - N_t(\omega)}$$

where $N_t(\omega)$ is the number of up-moves up to time $t$ along path $\omega$. Because $u \cdot d = 1$ in the CRR parametrisation, the tree **recombines**: an up-move followed by a down-move yields the same price as the reverse sequence. This reduces the number of distinct nodes from $2^n$ to $n+1$, making the model computationally efficient.

### 2.3 · Risk-Neutral Valuation

Under the no-arbitrage condition $d < 1+r < u$, the unique risk-neutral probability of an up-move is:

$$p = \frac{(1+r) - d}{u - d}, \qquad 1 - p = \frac{u - (1+r)}{u - d}$$

The fair price at $t = 0$ of any European contingent claim $X = D(S_n)$ is then:

$$\pi_0(X) = (1+r)^{-n} \, \mathbb{E}^{\mathbb{Q}}[X]$$

This is the **discrete risk-neutral valuation formula**, the analogue of the Black-Scholes formula in continuous time.

---
## 3 · Parametrisations of $u$ and $d$

The choice of $u$ and $d$ determines how accurately the binomial tree approximates the continuous-time dynamics of the underlying asset. Two standard parametrisations are used in this paper.

### 3.1 · Cox-Ross-Rubinstein (CRR) Parametrisation

The original CRR parametrisation sets:

$$u = e^{\sigma\sqrt{\Delta t}}, \qquad d = \frac{1}{u} = e^{-\sigma\sqrt{\Delta t}}$$

where $\sigma$ is the annualised volatility and $\Delta t = T/n$ is the length of each time step. By construction $ud = 1$, so the tree recombines exactly. This parametrisation does not explicitly incorporate the drift $r$; the risk-neutral probability $p$ handles discounting.

### 3.2 · Jarrow-Rudd (JR) Parametrisation

Jarrow and Rudd (1983) proposed an alternative that matches the first two moments of the log-normal distribution more directly:

$$u = e^{\left(r - \frac{\sigma^2}{2}\right)\Delta t + \sigma\sqrt{\Delta t}}, \qquad d = e^{\left(r - \frac{\sigma^2}{2}\right)\Delta t - \sigma\sqrt{\Delta t}}$$

This parametrisation is derived from the continuous geometric Brownian motion (GBM) equation by discretising the Itô dynamics $dS = rS\,dt + \sigma S\,dW$ over each interval $\Delta t$. As shown in Section 5, it is precisely the JR parametrisation that appears in the formal proof of convergence to Black-Scholes.

Both parametrisations satisfy the no-arbitrage condition $d < 1+r < u$ for sufficiently large $n$ and converge to the same Black-Scholes limit, though they exhibit different oscillation patterns for finite $n$.

## 4 · Option Pricing in the CRR Model

### 4.1 · European Call Option

The fair price at $t = 0$ of a European call with strike $K$ and maturity $n$ is given by the CRR call pricing formula:

$$C_0 = (1+r)^{-n} \sum_{j=0}^{n} \binom{n}{j} p^j (1-p)^{n-j} \max(S_0 u^j d^{n-j} - K,\, 0)$$

where $\tilde{p} = pu/(1+r)$. This can be written more compactly using the complementary binomial distribution $\Psi(\cdot\,;\,n,\,p)$.

### 4.2 · European Put via Put-Call Parity

Rather than deriving the put price directly, it is more elegant to apply the **put-call parity** relationship. For European options with the same strike $K$, maturity $T$, and underlying $S$:

$$C_t + K \cdot B(t, T) = P_t + S_t$$

where $B(t, T) = (1+r)^{-(T-t)}$ is the price of a zero-coupon bond. Solving for the put price:

$$P_0 = C_0 + K(1+r)^{-n} - S_0$$

Put-call parity is a **no-arbitrage condition**: any violation would create a riskless profit. In the CRR model, it holds exactly at every node.

### 4.3 · American Options and Early Exercise

An American option may be exercised at any time $t \leq n$. Its value must therefore satisfy:

$$V_{i,j} = \max\!\left(\text{intrinsic value},\; (1+r)^{-1}\left[p\,V_{i+1,j+1} + (1-p)\,V_{i+1,j}\right]\right)$$

The backward induction compares, at each node, the payoff from **immediate exercise** against the **continuation value** (the discounted expected value of holding the option). Because of this early-exercise premium, an American put is always worth at least as much as its European counterpart:

$$P_0^{\text{American}} \geq P_0^{\text{European}}$$

No closed-form solution exists for American options in general; the binomial tree is one of the standard numerical methods used in practice.

## 5 · Convergence to Black-Scholes

### 5.1 · The Black-Scholes Model

The Black-Scholes model is a continuous-time model in which the stock price follows a **geometric Brownian motion** under the risk-neutral measure $\mathbb{Q}$:

$$S_s = S_0 \, e^{\left(r - \frac{\sigma^2}{2}\right)s + \sigma W_s}, \qquad s \in [0, T]$$

where $(W_s)_{s \geq 0}$ is a standard Brownian motion under $\mathbb{Q}$. The bond price is $B_s = e^{rs}$.

The fair price at $t = 0$ of a European contingent claim $D(S_T)$ is:

$$D_0^{\text{BS}} = e^{-rT} \, \mathbb{E}^{\mathbb{Q}}\left[D(S_T)\right]$$

For a put option with payoff $\max(K - S_T, 0)$ this yields the celebrated Black-Scholes formula:

$$P_0^{\text{BS}} = K e^{-rT} \Phi(-d_2) - S_0 \Phi(-d_1)$$

where:

$$d_1 = \frac{\ln(S_0/K) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}, \qquad d_2 = d_1 - \sigma\sqrt{T}$$

and $\Phi$ is the standard normal CDF. The call price follows immediately from put-call parity:

$$C_0^{\text{BS}} = S_0 \Phi(d_1) - K e^{-rT} \Phi(d_2)$$

### 5.2 · The $n$-th CRR Model as a Discretisation

Divide $[0, T]$ into $n$ equal steps of length $\Delta t = T/n$. To align bond prices between the CRR model and the BS model at each real time $t \Delta t$, we require:

$$(1 + r_n)^t = e^{r t \Delta t} \implies r_n = e^{r \Delta t} - 1$$

With this choice, the JR parametrisation for $u$ and $d$ (Section 3.2) implies that the stock price in the $n$-th CRR model can be written as:

$$S_n^{(n)} = S\!\left(\frac{1}{\sqrt{n}} W_n^{(n)}\right), \qquad S(z) := S_0 \, e^{\left(r - \frac{\sigma^2}{2}\right)T + \sigma\sqrt{T}\,z}$$

where $W_n^{(n)} = \sum_{i=1}^n Y_i$ is a **symmetric random walk** with $Y_i = \pm\sqrt{\Delta t}$ (each with probability $\approx 1/2$ under $\mathbb{Q}_n$).

### 5.3 · No-Arbitrage Condition

The $n$-th CRR model is arbitrage-free if and only if $d_n < 1 + r_n < u_n$, which translates to:

$$n > \frac{1}{4\sigma^2 T}$$

Let $N$ be the smallest integer satisfying this inequality. For all $n \geq N$, the unique one-step risk-neutral probability is:

$$p_n = \frac{e^{\frac{1}{2}\sigma^2 \Delta t} - e^{-\sigma\sqrt{\Delta t}}}{e^{\sigma\sqrt{\Delta t}} - e^{-\sigma\sqrt{\Delta t}}}$$

### 5.4 · Main Theorem (Cutland and Roux, 2013)

**Theorem.** Let $D: \mathbb{R}_{>0} \to \mathbb{R}$ be a bounded continuous payoff function. For each $n \geq N$, define the path-independent derivative $D^{(n)} := D(S_n^{(n)})$ in the $n$-th CRR model, and let $D_0^{(n)}$ be its unique fair price at time $0$. Then:

$$\lim_{n \to \infty} D_0^{(n)} = e^{-rT} \, \mathbb{E}^{\mathbb{Q}}\left[D(S_T)\right]$$

That is:

$$\lim_{n \to \infty} (1+r_n)^{-n} \sum_{t=0}^{n} \binom{n}{t} p_n^t (1-p_n)^{n-t} D\!\left(S_0 u_n^t d_n^{n-t}\right) = e^{-rT} \int_{-\infty}^{+\infty} \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} D(S(z))\, dz$$

**Proof sketch.** Define $g(z) := D(S(z))$, which is continuous and bounded since $D$ is. The key step is the **weak convergence** of $\frac{1}{\sqrt{n}} W_n^{(n)}$ to a standard normal random variable under $\mathbb{Q}_n$, which follows from the CLT applied to the symmetric random walk $W_n^{(n)}$. Therefore:

$$\lim_{n\to\infty} \mathbb{E}^{\mathbb{Q}_n}\!\left[D(S_n^{(n)})\right] = \lim_{n\to\infty} \mathbb{E}^{\mathbb{Q}_n}\!\left[g\!\left(\tfrac{1}{\sqrt{n}} W_n^{(n)}\right)\right] = \mathbb{E}^{\mathbb{Q}}[g(Z)] = \mathbb{E}^{\mathbb{Q}}[D(S_T)]$$

Combined with $(1+r_n)^n = e^{rT}$, this gives the result. $\blacksquare$

### 5.5 · Convergence of Put and Call Prices

The put payoff $\max(K - S_T, 0)$ is bounded and continuous (extended to $K$ for $S_T \leq 0$), so the theorem applies directly:

$$\lim_{n\to\infty} P_0^{(n)} = P_0^{\text{BS}} = e^{-rT} K\Phi(-d_2) - S_0\Phi(-d_1)$$

The call payoff $\max(S_T - K, 0)$ is **unbounded**, so the theorem cannot be applied directly. However, using put-call parity in the $n$-th CRR model:

$$C_0^{(n)} = P_0^{(n)} + S_0 - (1+r_n)^{-n} K$$

Since $(1+r_n)^{-n} \to e^{-rT}$ and $P_0^{(n)} \to P_0^{\text{BS}}$, it follows that $C_0^{(n)} \to C_0^{\text{BS}}$.

## 7 · Conclusions

This paper has presented the Cox-Ross-Rubinstein binomial model as a rigorous discrete-time framework for option pricing, and demonstrated its convergence to the Black-Scholes continuous-time model.

The key results are:

1. **Arbitrage-free pricing.** Under the condition $d < 1 + r < u$, the CRR market is complete and arbitrage-free, yielding a unique risk-neutral measure and a unique fair price for every European contingent claim.

2. **Convergence.** As $n \to \infty$, CRR prices converge to Black-Scholes prices for any bounded continuous payoff. The proof relies on the weak convergence of the normalised random walk $\frac{1}{\sqrt{n}} W_n^{(n)}$ to a standard normal random variable — a consequence of the Central Limit Theorem.

3. **Numerical behaviour.** The convergence rate is approximately $O(1/n)$, with a characteristic even/odd oscillation that reflects the discrete nature of the binomial tree. Both CRR and JR parametrisations converge to the same limit, though with different oscillation patterns.

4. **American options.** The binomial framework naturally extends to American options via backward induction with early-exercise comparison, a feature not available in the Black-Scholes closed form.

The CRR model thus serves a dual purpose: it is a practical numerical method for pricing derivatives (especially American options), and a theoretical bridge between discrete and continuous time — showing that the Black-Scholes formula is not an assumption, but a limit.
